Demonstrar a execução do cálculo e utilizar um comentário para afirmar que é um programa de exemplo de folha de pagamento.

```java
import java.util.Scanner;
Scanner scan = new Scanner(System.in);
System.out.print("Informe um inteiro: ");
int b = scan.nextInt();
System.out.print("Informe um inteiro: ");
int c = scan.nextInt();

int a = b * c;

System.out.printf("O resultado da operação %d x %d é %d", b, c, a);
scan.close();
// programa executa um cálculo de exemplo de folha de pagamento
```
